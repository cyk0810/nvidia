```python
# YOLOv8 설치
!pip install ultralytics --quiet

# 라이브러리 임포트
from ultralytics import YOLO
from PIL import Image, ImageDraw, ImageFont
from google.colab import files
from IPython.display import Image as IPyImage, display
import os
import cv2
import numpy as np

# YOLOv8 모델 로드
model = YOLO('yolov10x.pt')

# 이미지 업로드
uploaded = files.upload()

# 출력 폴더 생성
os.makedirs("outputs", exist_ok=True)

# 이미지 순회 처리
for img_path in uploaded.keys():
    results = model(img_path)
    result = results[0]
    boxes = result.boxes

    if boxes is None or len(boxes.cls) == 0:
        print(f"'{img_path}'에서는 객체가 감지되지 않아 제외됨")
        continue

    # 클래스별 객체 수 계산
    names = model.names
    class_counts = {}
    for cls_id in boxes.cls:
        cls_name = names[int(cls_id)]
        class_counts[cls_name] = class_counts.get(cls_name, 0) + 1

    # YOLO 탐지 결과 이미지(박스와 라벨 그려진 이미지)를 numpy로 받기
    annotated_img = result.plot()  # numpy array (BGR)

    # PIL 이미지로 변환
    image = Image.fromarray(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()

    # 객체 수 요약 텍스트 생성
    text_lines = [f"{name}: {count}" for name, count in class_counts.items()]
    line_sizes = [draw.textbbox((0, 0), line, font=font) for line in text_lines]
    text_width = max(x1 - x0 for (x0, y0, x1, y1) in line_sizes)
    text_height = sum(y1 - y0 for (x0, y0, x1, y1) in line_sizes)

    # 텍스트 박스 위치
    margin = 10
    box_x = image.width - text_width - margin
    box_y = margin
    draw.rectangle(
        [box_x - 5, box_y - 5, box_x + text_width + 5, box_y + text_height + 5],
        fill="black"
    )

    # 텍스트 작성
    current_y = box_y
    for line, (x0, y0, x1, y1) in zip(text_lines, line_sizes):
        draw.text((box_x, current_y), line, fill="white", font=font)
        current_y += (y1 - y0)

    # 저장 및 출력
    output_path = f"outputs/summary_{img_path}"
    image.save(output_path)
    display(IPyImage(output_path))
```
